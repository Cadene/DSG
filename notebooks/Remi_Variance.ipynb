```python
def add_dummies(frames, X, col):
    frames.append(pd.get_dummies(X[col], prefix=col))

cols_categorical = [
    'SelectedPackage',
    'FirstDriverMaritalStatus',
    'CarFuelId',
    'CarUsageId',
    'CarParkingTypeId',
    'FirstDriverDrivingLicenceType',
    'CoverIsNoClaimDiscountSelected',
    'CarDrivingEntitlement',
    'CarTransmissionId',
    'AllDriversNbConvictions',
    'IsPolicyholderAHomeowner',
    'NameOfPolicyProduct'
] 

cols_continuous = [
        'CarAnnualMileage',
        'FirstDriverAge',
        'CarInsuredValue',
        'CarAge',
        'VoluntaryExcess',
        'PolicyHolderNoClaimDiscountYears',
        'AllDriversNbConvictions',
        'DaysSinceCarPurchase'
]

cols_left = [
        'SCID',
        'FirstDriverDrivingLicenseNumberY',
        'SocioDemographicId',
        'PolicyHolderResidencyArea',
        'RatedDriverNumber',
        'CarMakeId',
        'AffinityCodeId'
]

def extract_features(X, cols_categorical, cols_continuous):
    frames = []
    for col in cols_categorical:
        add_dummies(frames, X, col)
    for col in cols_continuous:
        frames.append(X[col])
    return pd.concat(frames, axis=1)
```

```python
X_train4 = extract_features(X_train3, cols_categorical, cols_continuous)
len(X_train4.columns)
```

```python
X_test2 = extract_features(X_test, cols_categorical, cols_continuous)
len(X_test2.columns) # X_test
```
